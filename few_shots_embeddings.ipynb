```python
# embeded_docs
text_splitter=CharacterTextSplitter.from_tiktoken_encoder(chunk_size=270) # split the text in tiktoken using CharacterTextSplitter 
docs=text_splitter.split_documents(data) #split the text
```

```python
db=Chroma.from_documents(docs,embeddings,persist_directory="./few_show") # create the  database as few_show
db.persist()
```
